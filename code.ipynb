### 1.Logistics  drop

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
Train_logistics,Test_logistics= train_test_split(df,shuffle=True) #Default test = 25%
```

```python
Train_logistics,Test_logistics= train_test_split(df,shuffle=True)
from sklearn.model_selection import GridSearchCV
#  solver & C
param_grid = {'C': [1, 10, 100], 'solver': ['liblinear', 'lbfgs', 'newton-cg', 'sag', 'saga']}
logistics_model = LogisticRegression()
#  GridSearchCV  Hyperparameter 
grid_search_log = GridSearchCV(logistics_model, param_grid, cv=5)
grid_search_log.fit(Train_logistics[["Age",'Sex','Chest Pain Type', 'Resting Blood Pressure',
                                     'Cholestrol', 'Fasting Blood Sugar','Resting Electrocardiographic Results',
                                     'Maximum Heart Rate Achieved', 'Exercise Induced Angina',
                                     'ST Depression Induced', 'Slope of Peak ST Segment', 
                                     'Number of Major Blood Vessels', 'Type']],
                                      Train_logistics['Target'])
#  kernel & C 
print("Best Kernel:", grid_search_log.best_params_['solver'])
print("Best C:", grid_search_log.best_params_['C'])
```

```python
##classifier_logistics = LogisticRegression(C=10, solver='liblinear',random_state=0) # model logistics
```

```python
#  Train
classifier_logistics = LogisticRegression(C=10, solver='liblinear',random_state=0) # model logistics
classifier_logistics.fit(Train_logistics[["Age",'Sex','Chest Pain Type', 'Resting Blood Pressure',
                                     'Cholestrol', 'Fasting Blood Sugar','Resting Electrocardiographic Results',
                                     'Maximum Heart Rate Achieved', 'Exercise Induced Angina',
                                     'ST Depression Induced', 'Slope of Peak ST Segment', 
                                     'Number of Major Blood Vessels', 'Type']],
                                      Train_logistics['Target'])
```

```python
y_pred_test_logistics = classifier_logistics.predict(Test_logistics[["Age",'Sex','Chest Pain Type', 'Resting Blood Pressure',
                                       'Cholestrol', 'Fasting Blood Sugar', 'Resting Electrocardiographic Results',
                                       'Maximum Heart Rate Achieved', 'Exercise Induced Angina',
                                       'ST Depression Induced', 'Slope of Peak ST Segment',
                                       'Number of Major Blood Vessels', 'Type']])
```

```python
from sklearn.metrics import precision_score,accuracy_score,recall_score,f1_score,classification_report, confusion_matrix
precision = precision_score(Test_logistics["Target"], y_pred_test_logistics)
accuracy = accuracy_score(Test_logistics["Target"], y_pred_test_logistics)
recall=recall_score(Test_logistics["Target"], y_pred_test_logistics)
f1=f1_score(Test_logistics["Target"], y_pred_test_logistics)
ClassificationReport=classification_report(Test_logistics["Target"], y_pred_test_logistics)
print("\nAccuracy: ",accuracy)
print("\nPrecision: ",precision)
print("\nRecall: ",recall)
print("\nF1 Score: ",f1)
print("\nConfusion Matrix: \n")
print(confusion_matrix(Test_logistics.Target, y_pred_test_logistics))
print("\nClassification Report: \n")
print(ClassificationReport)
```

```python
classifier_svm = SVC(C=100, kernel = 'linear', random_state = 0)
classifier_svm.fit(Train_svm[["Age",'Sex','Chest Pain Type','Resting Blood Pressure','Cholestrol',
                          'Fasting Blood Sugar','Resting Electrocardiographic Results','Maximum Heart Rate Achieved',
                          'Exercise Induced Angina','ST Depression Induced','Slope of Peak ST Segment',
                          'Number of Major Blood Vessels','Type']],Train_svm['Target'])
```

```python
y_pred_test_svm = classifier_svm.predict(Test_svm[["Age",'Sex','Chest Pain Type','Resting Blood Pressure','Cholestrol',
                                               'Fasting Blood Sugar','Resting Electrocardiographic Results',
                                               'Maximum Heart Rate Achieved','Exercise Induced Angina','ST Depression Induced',
                                               'Slope of Peak ST Segment','Number of Major Blood Vessels','Type']])
```

```python
precision = precision_score(Test_svm["Target"], y_pred_test_svm)
accuracy = accuracy_score(Test_svm["Target"], y_pred_test_svm)
recall=recall_score(Test_svm["Target"], y_pred_test_svm)
f1=f1_score(Test_svm["Target"], y_pred_test_svm)
ClassificationReport=classification_report(Test_svm["Target"], y_pred_test_svm)
print("\nAccuracy: ",accuracy)
print("\nPrecision: ",precision)
print("\nRecall: ",recall)
print("\nF1 Score: ",f1)
print("\nConfusion Matrix: \n")
print(confusion_matrix(Test_svm.Target, y_pred_test_svm))
print("\nClassification Report: \n")
print(ClassificationReport)
```
